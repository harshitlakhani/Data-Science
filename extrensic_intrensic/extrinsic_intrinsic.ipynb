```python
import os
import json
import pandas as pd
```

```python
path = 'C:/Users/USER/OneDrive/Desktop/All JN Projects/Sigma Apu/New folder/Denchmark_BRs-main/JSonSet/simplejson'
```

```python
folders = os.listdir(path)
```

```python
BR_id = []
BRsummary = []
BRdescription = []
Project_name = []
```

```python
for folder in folders:
    file_path=path+"/"+folder
    files = os.listdir(file_path)
    for file in files:
        full_path=path+"/"+folder+"/"+file
        with open(full_path) as f:
            data = json.load(f)
        real = data["BR"]
        BR_id.append(real["BR_id"])
        BRsummary.append(real["BR_text"]["BRsummary"])
        BRdescription.append(real["BR_text"]["BRdescription"])
        Project_name.append(folder)
```

```python
df = pd.DataFrame(list(zip(BR_id, BRsummary, BRdescription, Project_name)),columns =['BR_id', 'BRsummary', 'BRdescription'])
```
